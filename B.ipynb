```python
T_list_1=[]
T_list_2=[]
for i in df1_2['时间 (Time)']:
    label=str(i)
    T1=df1_1[df1_1['时间 (Time)']==label]['系统I温度 (Temperature of system I)'].values
    T2=df1_1[df1_1['时间 (Time)']==label]['系统II温度 (Temperature of system II)'].values

    T_list_1.append(T1)
    T_list_2.append(T2)
T_list_1_2=[]
T_list_2_2=[]

for i in T_list_1:
    for j in i:
        T_list_1_2.append(j)
for i in T_list_2:
    for j in i:
        T_list_2_2.append(j)

print(len(T_list_1))
print(len(T_list_2))
```
